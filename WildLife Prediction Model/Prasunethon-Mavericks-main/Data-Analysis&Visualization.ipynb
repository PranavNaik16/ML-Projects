# Statistical Analysis

## Range of Dataset

```python
for col in data.columns:
    print(col, ":", data[col].min(), "-", data[col].max())
```

## Standard Deviation

```python
for col in data.select_dtypes(include = "number").columns:
    print(col, ":", data[col].std())
```

## Mean

```python
for col in data.select_dtypes(include = "number").columns:
    print(col, ":", data[col].mean())
```

# Median

```python
for col in data.select_dtypes(include = "number").columns:
    print(col, ":", data[col].median())
```

# Variance

```python
for col in data.select_dtypes(include = "number").columns:
    print(col, ":", data[col].var())
```

## Correlation

```python
data.select_dtypes(include = "number").corr()
```

## Covariance

```python
data.select_dtypes(include = "number").cov()
```

```python
plt.figure(figsize=(12, 5))

# Plot each temperature type separately
plt.plot(data["year_month"], data["average_temperature_median"], label = 'Average Temperature')
plt.plot(data["year_month"], data["lat_lag_median"], label = 'Latitude Lag')
plt.plot(data["year_month"], data["lon_lag_median"], label = 'Longitude Lag')

plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))  # Increased height to make room for more plots

plt.plot(data["year_month"], data["precipitation_lag_median"], label='Precipitation Lag Median')
plt.plot(data["year_month"], data["wind_speed_lag_median"], label='Wind Speed Lag Median')
plt.plot(data["year_month"], data["maximum_sustained_wind_speed_lag_median"], label='Max Sustained Wind Speed Lag Median')
plt.plot(data["year_month"], data["fog_lag_mean"], label='Fog Lag Mean')
plt.plot(data["year_month"], data["thunder_lag_mean"], label='Thunder Lag Mean')

plt.xticks(rotation='vertical')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position to avoid overlapping
plt.tight_layout()  # Adjust the layout to make room for the legend
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plot dew point, precipitation, and wind gust
plt.plot(data["year_month"], data["dew_point_lag_median"], label='Dew Point Lag Median')
plt.plot(data["year_month"], data["snow_depth_lag_median"], label='Snow Depth Lag Median')
plt.plot(data["year_month"], data["wind_gust_lag_median"], label='Wind Gust Lag Median')

plt.xticks(rotation='vertical')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position to avoid overlapping
plt.tight_layout()  # Adjust the layout to make room for the legend
plt.show()
```

```python
#sns.set() sets the aesthetic style of the plots
sns.set()
fig, axes = plt.subplots(2, 2, figsize = (15, 10))
plt.tight_layout(pad = 5.0)
# Plot 1: Countplot of 'Survived'
sns.distplot(data["average_temperature_median"], ax = axes[0, 0])             
axes[0, 0].set_title("average_temperature_median")
# Plot 2: Countplot of 'Sex'
sns.distplot(data["maximum_temperature_median"], ax = axes[0, 1])
axes[0, 1].set_title("maximum_temperature_median")
# Plot 3: Countplot of 'Pclass'
sns.distplot(data["minimum_temperature_median"], ax = axes[1, 0])
axes[1, 0].set_title("minimum_temperature_median")

# Display the plots
plt.show()
```

```python
filtered_df = data[data['is_fire'] == 0]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['year_month'], filtered_df['lat_lag_median'], marker='o')

# Add labels and title
plt.xlabel('Year-Month')
plt.ylabel('Lat Lag Median')
plt.title('Lat Lag Median vs Year-Month (is_fire = 0)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
filtered_df = data[data['is_fire'] == 0]

# Plot the data
plt.figure(figsize=(14, 8))

# List of columns to plot
columns_to_plot = [
    'average_temperature_median', 'maximum_temperature_median', 'minimum_temperature_median',
    'wind_speed_lag_median', 'maximum_sustained_wind_speed_lag_median', 'wind_gust_lag_median',
    'dew_point_lag_median', 'precipitation_lag_median'
]

# Plot each column
for column in columns_to_plot:
    plt.plot(filtered_df['year_month'], filtered_df[column], marker='o', label=column)

# Add labels and title
plt.xlabel('Year-Month')
plt.ylabel('Values')
plt.title('No Wild Fire')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot example: average temperature vs. wind speed
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['average_temperature_median'], filtered_df['wind_speed_lag_median'])
plt.xlabel('Average Temperature Median')
plt.ylabel('Wind Speed Lag Median')
plt.title('Average Temperature vs Wind Speed (is_fire = 0)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
