# Feature Extraction
In this module, we will try to reduce the number of features.  
1. Club the dataset on the basis of months by taking mean.
2. Hence we will have the data of 120 months.
3. Further, there are 15 lags provided for the same feature.
4. Hence, we shall combine those 15 lags into 1 on the basis of median.
5. This is done to reduce the complexity of the problem and to aid in Data Analysis and Visualization.

```python
# Convert the date column to datetime format
data['acq_date'] = pd.to_datetime(data['acq_date'])

# Extract year and month
data['year'] = data['acq_date'].dt.year
data['month'] = data['acq_date'].dt.month

# Group by year and month, then calculate the mean for each group
monthly_means = data.groupby(['year', 'month']).mean().reset_index()

# Create a new dataframe where columns represent the months
months = []
for year in range(2013, 2023):
    for month in range(1, 13):
        months.append(f"{year}-{month:02d}")

monthly_means['year_month'] = monthly_means['year'].astype(str) + '-' + monthly_means['month'].astype(str).str.zfill(2)

# Drop the original year and month columns
monthly_means = monthly_means.drop(columns=['year', 'month'])

# Set 'year_month' as the index
monthly_means.set_index('year_month', inplace=True)

# Sort the index to ensure the DataFrame is in the correct order
monthly_means = monthly_means.reindex(months)
```
